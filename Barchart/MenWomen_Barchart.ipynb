```python
import numpy as np
import matplotlib.pyplot as plt

groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)

# create plot
index = np.arange(groups)
bar_width = 0.35

rects1 = plt.bar(index, men_means, bar_width, color='green', label='Men')

rects2 = plt.bar(index + bar_width, women_means, bar_width, color='red',label='Women')

plt.legend()

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
```
