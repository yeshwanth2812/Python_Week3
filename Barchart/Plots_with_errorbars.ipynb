```python
import numpy as np
import matplotlib.pyplot as plt
group = 5
menMeans = (54.74, 42.35, 67.37, 58.24, 30.25)
menStd = (4, 3, 4, 1, 5)
# the x locations for the groups
index = np.arange(group)    
# the width of the bars
width = 0.35   
plt.bar(index, menMeans, width, menStd)
plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity')
```
