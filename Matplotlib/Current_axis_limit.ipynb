```python
import matplotlib.pyplot as plt

x1 = [10,20,30]
y1 = [30,20,10]

plt.plot(x1, y1, label = 'line')

# shows the current axis limits values
print(plt.axis()) 
# Limit of x axis 0 to 100
# Limit of y axis 0 to 200
plt.axis([0, 100, 0, 200]) 

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
```
